```python
for H in Ht:
    height_date = H
    height_str = height_date.strftime('%Y%m%d%H%M%S')
    height_int = int(height_str)
    
    if height_int > 20230000000000:
        height = (height_int/1000000) - 20230000
    else:
        height = (height_int/1000000) - 20000000
        
    height_ft_str = str(height)[0]
    height_ft = int(height_ft_str)
    height_in = height - height_ft * 100
    height = height_ft * 12 + height_in
    Ht_in.append(height)
```

```python
Draft = CombineResults['Drafted (tm/rnd/yr)']
```

```python
Draft
```

```python
type(Draft[4])
```

```python
Draft_Pick = []
```

```python
for n in Draft:
    
    if pd.isnull(n) == True:
        Draft_Pick.append(np.nan)
        
    else:
        Split_String = n.split()
        
        if len(Split_String) == 9:
            Pick = Split_String[5]
        else:
            Pick = Split_String[6]
        
        Pick = Pick.replace("s", "")
        Pick = Pick.replace("t", "")
        Pick = Pick.replace("h", "")
        Pick = Pick.replace("d", "")
        Pick = Pick.replace("n", "")
        Pick = Pick.replace("r", "")
        
        
        Pick_int = int(Pick)
        
        Draft_Pick.append(Pick_int)
```

```python
Draft_Pick
```

```python
CombineResults['Draft_Pick'] = Draft_Pick
```
